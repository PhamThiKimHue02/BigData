```python
import os
import shutil

# Tạo thư mục đích
os.makedirs("./data/mvtec", exist_ok=True)

# Đường dẫn dataset Kaggle đã mount
src_path = "/kaggle/input/mvtec-ad"

# Copy toàn bộ dataset vào ./data/mvtec
shutil.copytree(src_path, "./data/mvtec", dirs_exist_ok=True)

# Kiểm tra cấu trúc
for root, dirs, files in os.walk("./data/mvtec"):
    level = root.replace("./data/mvtec", "").count(os.sep)
    indent = " " * 4 * (level)
    print(f"{indent}{os.path.basename(root)}/")
    subindent = " " * 4 * (level + 1)
    for f in files:
        print(f"{subindent}{f}")
```
